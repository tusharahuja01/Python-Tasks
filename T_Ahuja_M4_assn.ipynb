# Task - 1

What proportion of customers have an age of more than the average customer age but less than 5

### Comments

* The code reads the data from the provided GitHub repository link and loads it into a Pandas dataframe named 'df'. After that, it calculates the average age of the customers in the dataset using the 'mean' function on the 'age' column of the dataframe.

* Next, it filters the dataframe 'df' to include only the customers with an age greater than the average customer age (calculated in the previous step) but less than 56. It then calculates the proportion of these customers over the total number of customers in the dataset.

* Finally, it prints the proportion of the customers with an age greater than the average customer age but less than 56, rounding the result to two decimal places.

```python
import pandas as pd

# Read the data from the GitHub repository into a Pandas dataframe
github_url = 'https://raw.githubusercontent.com/tusharahuja01/DAV-5400/main/M4_Data.csv'
df = pd.read_csv(github_url)

# Calculate the average customer age
average_age = df['age'].mean()

# Calculate the proportion of customers with an age greater than the average customer age but less than 5
customers = df[(df['age'] > average_age) & (df['age'] < 56)]
proportion = len(customers) / len(df)

print('Proportion of customers with an age of more than the average customer age but less than 56 are:', proportion)
```

# Task - 2

How many customers that have purchased Product B have spent more on Product B than 1.5x the
median amount that customers typically spend on Product B?

### Comments

* To calculate the median amount that customers typically spend on Product B, I used the median function on the 'turnover_B' column in the dataframe 'df'.

* Next, I filtered customers who have purchased Product B and spent more than 1.5 times the median amount on Product B. This was done by using boolean indexing in pandas to create a subset of the dataframe that met both conditions.

* Finally, I counted the number of customers in the subset and stored the total count in the variable 'total_customers'. I then printed out the number of customers that have purchased Product B and spent more than 1.5 times the median amount on Product B.

```python
# Calculate the median amount that customers typically spend on Product B
median_product_b = df['turnover_B'].median()

# Filter the customers that have purchased Product B and have spent more than 1.5x the median amount
customers_spent_more = df[(df['prod_B'] == 1) & (df['turnover_B'] > 1.5 * median_product_b)]

# Count the number of customers
total_customers = len(customers_spent_more)

print("Number of customers that have purchased Product B and spent more than 1.5x the median amount on Product B:", total_customers)
```

# Task - 3

How many customers have spent more on Product B than they have on Product A?

### Comments

* To filter the DataFrame for customers who have spent more on Product B than on Product A, we first create a new DataFrame called `customers` by using boolean indexing. We compare the 'turnover_B' column with the 'turnover_A' column to check which customers have spent more on Product B than on Product A. Those customers are selected and stored in the new DataFrame.

* Next, we count the number of customers who meet the criteria by using the `len()` function on the `customers` DataFrame. This gives us the total number of customers who have spent more on Product B than on Product A.

* Finally, we print out the total number of customers who have spent more on Product B than on Product A using the formatted string "Number of customers who have spent more on Product B than on Product A:" followed by the `total_customers` variable.

```python
# Filter the DataFrame for customers who have spent more on Product B than on Product A
customers = df[df['turnover_B'] > df['turnover_A']]

# Count the number of customers who meet the criteria
total_customers = len(customers)

print("Number of customers who have spent more on Product B than on Product A:", total_customers)
```

# Task - 4

How many customers have purchased either type 6 of Product A or type 9 of Product?

### Comments

* To count the number of customers who have purchased type 6 of Product A or type 9 of Product B, I used a conditional statement within the DataFrame `df` to filter rows where `type_A` is 6 or `type_B` is 9. I then used the `nunique()` method on the 'ID' column to calculate the total unique customers who meet this condition. The result is stored in the variable `total_customers` and printed as the output.

```python
# Count the number of customers who have purchased type 6 of Product A or type 9 of Product B
total_customers = df[(df['type_A'] == 6) | (df['type_B'] == 9)]['ID'].nunique()

print("Number of customers who have purchased either type 6 of Product A or type 9 of Product B:", total_customers)
```

# Task - 5

We are provided with the number of months for which each customer has been a client of the
company (lor_M). Which customer has attained the highest average amount of monthly spending (i.e., across all products they have purchased from the company) during the time in which they have been a customer of the company? Answer by providing the dataframe row index, customer ID, and the total number of months the customer has been with the company, as well as that customer's total spending amount. Finally, display the complete content of the dataframe row corresponding to the customer you have identified.

### Comments

* I have approached the task by first calculating the total spending amount for each customer by adding their turnover from both A and B. Then, I calculated the average monthly spending for each customer by dividing their total spending with the length of the relationship in months. 

* Next, I identified the customer with the highest average monthly spending by finding the row with the maximum value in the 'Avg_Monthly_Spending' column. I displayed the required information which includes the customer ID, total number of months spent with the company, total spending amount, and average monthly spending for that customer.

* Finally, I displayed the complete content of the dataframe row corresponding to the identified customer to provide an overview of all their details.



```python
# Calculate the total spending amount for each customer
df['Total_Spending'] = df['turnover_A'] + df['turnover_B']

# Calculate the average monthly spending for each customer
df['Avg_Monthly_Spending'] = df['Total_Spending'] / df['lor_M']

# Identify the customer with the highest average monthly spending
customer_highest_avg_spending = df.loc[df['Avg_Monthly_Spending'].idxmax()]

# Display the required information
print("Customer ID:", customer_highest_avg_spending['ID'])
print("Total Number of Months with the Company:", customer_highest_avg_spending['lor_M'])
print("Total Spending Amount:", customer_highest_avg_spending['Total_Spending'])
print("Average Monthly Spending:", customer_highest_avg_spending['Avg_Monthly_Spending'])

# Display the complete content of the dataframe row corresponding to the identified customer
print("\nCustomer Information:")
print(customer_highest_avg_spending)
```
